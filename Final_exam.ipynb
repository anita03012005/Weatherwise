```python
# Section 2 – Manual corrected script (Week 10 only)

# In-memory list for tasks
tasks = []

def add_task(task):
    """
    Add a task and return a status string.
    This enables simple doctests.

    >>> tasks.clear(); add_task("Buy milk")
    'Added: Buy milk'
    """
    tasks.append(task)
    msg = f"Added: {task}"
    print(msg)
    return msg

def remove_task(index):
    """
    Remove a task by zero-based index.
    Fixes:
    - Typo 'tas' -> 'tasks'
    - Added lower-bound check (0 <= index)

    >>> tasks.clear(); _ = add_task("A"); _ = add_task("B"); remove_task(1)
    'Removed: B'
    >>> remove_task(5)
    'Invalid index'
    >>> remove_task(-1)
    'Invalid index'
    """
    if 0 <= index < len(tasks):
        removed = tasks[index]
        del tasks[index]
        msg = f"Removed: {removed}"
    else:
        msg = "Invalid index"
    print(msg)
    return msg

def list_tasks():
    """
    Show all tasks using zero-based indexes.
    Fix: enumerate the correct name 'tasks', not 'task'.
    Return a single string so doctest can assert.

    >>> tasks.clear(); _ = add_task("X"); _ = add_task("Y"); list_tasks()
    '0: X\\n1: Y'
    >>> tasks.clear(); list_tasks()
    'No tasks'
    """
    if not tasks:
        msg = "No tasks"
    else:
        lines = [f"{i}: {t}" for i, t in enumerate(tasks)]
        msg = "\n".join(lines)
    print(msg)
    return msg

def main():
    add_task("Buy milk")
    add_task("Pay bills")
    add_task("Walk dog")
    list_tasks()
    remove_task(1)
    list_tasks()

if __name__ == "__main__":
    import doctest
    doctest.testmod()
    main()
```

# Section 3.2 – Function Refinement with Course Constraints

Prompt to AI:
Please help me rewrite safe_weather_data_fetch(city) using Week 8 only. Use one simple try/except, input validation with if statements, short print messages for errors, and return either a dictionary or None. Add doctest examples that match the Week 9 style. Avoid specific exception types and multiple except blocks.

Course References:
Error handling approach: Week 8 basic try/except, simple prints.
Testing approach: Week 9 doctest examples placed in the function docstring.
Defensive checks: Week 7 verify structure before using it.


```python
import requests

def refined_safe_weather_data_fetch(city):
    """
    Fetch weather data with basic error handling - Week 8 style.
    Uses simple input checks, one try/except, and clear prints.
    Returns dict on success, or None on failure.

    >>> refined_safe_weather_data_fetch("")
    Error: City name cannot be empty
    >>> refined_safe_weather_data_fetch("InvalidCity123")  # doctest: +SKIP
    Error: Could not connect to weather service
    """
    # Week 6: simple input validation
    if not city:
        print("Error: City name cannot be empty")
        return None

    try:
        # Build URL
        url = f"http://wttr.in/{city}?format=j1"

        # Network request
        response = requests.get(url)
        data = response.json()

        # Week 7: defensive checks on shape before access
        # Expect keys and a non-empty list for current_condition
        if not isinstance(data, dict):
            print("Error: Unexpected response format")
            return None
        if "current_condition" not in data:
            print("Error: Missing 'current_condition' in response")
            return None
        if not data["current_condition"]:
            print("Error: No current weather data found")
            return None

        current = data["current_condition"][0]

        # Check minimal required fields exist
        if "temp_C" not in current or "windspeedKmph" not in current:
            print("Error: Missing temperature or wind fields")
            return None
        if "weatherDesc" not in current or not current["weatherDesc"]:
            print("Error: Missing weather description")
            return None

        # Build result dict (Week 8: simple dict)
        weather_info = {
            "city": city,
            "temperature": current["temp_C"],
            "wind_speed": current["windspeedKmph"],
            "description": current["weatherDesc"][0].get("value", "")
        }
        return weather_info

    except:
        # Week 8: single bare except with a short, clear message
        print("Error: Could not connect to weather service")
        return None
```

```python
def display_menu():
    """
    Display menu options using simple prints and also return the menu text.
    Week 3 formatting and Week 5 menu pattern.

    >>> # Printed menu (manual check)
    >>> display_menu()  # doctest: +SKIP
    1. Add task
    2. List tasks
    3. Remove task
    4. Quit

    >>> # Programmatic check of returned string
    >>> menu_text = display_menu()
    >>> "1. Add task" in menu_text and "4. Quit" in menu_text
    True
    """
    lines = [
        "1. Add task",
        "2. List tasks",
        "3. Remove task",
        "4. Quit"
    ]
    text = "\n".join(lines)
    print(text)
    return text


def get_user_choice():
    """
    Get and validate user choice in Week 6 input validation style.
    Accept only '1', '2', '3', '4'. Keep asking until valid.

    >>> # Interactive function. Use manual test.
    >>> # get_user_choice()  # doctest: +SKIP
    """
    choice = input("Choose 1, 2, 3, or 4: ").strip()
    while choice not in {"1", "2", "3", "4"}:
        print("Please enter 1, 2, 3, or 4.")
        choice = input("Choose 1, 2, 3, or 4: ").strip()
    return choice


def main():
    """
    Main program loop using Week 5 while loop pattern only.
    Stores tasks in a simple list. Uses basic prints and inputs.
    No doctest needed here.
    """
    tasks = []
    print("Simple Task Manager (Week 4 to Week 6 level)")
    running = True

    while running:
        display_menu()
        choice = get_user_choice()

        if choice == "1":
            title = input("Enter task title: ").strip()
            if title == "":
                print("Title cannot be empty.")
            else:
                tasks.append(title)
                print("Task added.")
        elif choice == "2":
            if not tasks:
                print("No tasks.")
            else:
                print("Tasks:")
                for i, t in enumerate(tasks):
                    print(f"{i}: {t}")
        elif choice == "3":
            if not tasks:
                print("No tasks to remove.")
            else:
                print("Current tasks:")
                for i, t in enumerate(tasks):
                    print(f"{i}: {t}")
                idx = input("Enter index to remove: ").strip()
                # Week 6 style validation
                if idx.isdigit():
                    i = int(idx)
                    if 0 <= i < len(tasks):
                        removed = tasks.pop(i)
                        print(f"Removed: {removed}")
                    else:
                        print("Invalid index.")
                else:
                    print("Please enter a number.")
        else:  # choice == "4"
            print("Goodbye.")
            running = False


if __name__ == "__main__":
    # Doctest runner as per Week 9 habit
    import doctest
    doctest.testmod(verbose=True)
```
